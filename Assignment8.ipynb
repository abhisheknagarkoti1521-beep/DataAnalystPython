Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

Answer:

**Simple Linear Regression (SLR)** is a statistical technique used to study the **linear relationship between two variables**, where one variable is **independent (X)** and the other is **dependent (Y)**. It assumes that the change in the dependent variable can be explained by the change in the independent variable through a straight-line relationship.

The general mathematical equation of Simple Linear Regression is:

Y = a + bX

Where:

* **Y** = Dependent variable (output/result)
* **X** = Independent variable (input/cause)
* **a** = Intercept (value of Y when X = 0)
* **b** = Regression coefficient or slope (rate of change of Y with respect to X)

### Purpose of Simple Linear Regression

1. **To establish a relationship between variables**

   SLR helps in understanding how one variable influences another. For example, it shows how sales depend on advertising expenditure.

2. **To predict future values**

   Once the regression equation is formed, it can be used to predict the value of the dependent variable for a given independent variable.

3. **To measure the strength and direction of the relationship**

   The slope indicates whether the relationship is positive or negative and how strongly X affects Y.

4. **To analyze trends and patterns**

   SLR is useful in identifying trends in data over time, such as population growth or demand forecasting.

5. **To support decision making**

   It is widely used in business, economics, engineering, and social sciences to make informed decisions based on data.

### Example

If **X represents the number of hours studied** and **Y represents marks obtained**, Simple Linear Regression helps predict marks based on the number of study hours.



Question 2: What are the key assumptions of Simple Linear Regression?

Answer:

Simple Linear Regression works properly only when certain **assumptions** about the data are satisfied. These assumptions ensure that the results obtained from the regression model are **valid, reliable, and accurate**.

### **1. Linearity**

There must be a **linear relationship** between the independent variable (X) and the dependent variable (Y).
This means that the change in Y should be directly proportional to the change in X and can be represented by a straight line.

### **2. Independence of Errors**

The **errors (residuals)** should be independent of each other.
This means that the error for one observation should not affect or be related to the error of another observation.

### **3. Homoscedasticity (Constant Variance)**

The variance of the errors should be **constant** for all values of X.
In other words, the spread of residuals should remain the same across the regression line.

### **4. Normality of Errors**

The residuals should be **normally distributed** with a mean of zero.
This assumption is important for hypothesis testing and confidence intervals.

### **5. No Multicollinearity**

In Simple Linear Regression, there should be **only one independent variable**.
Therefore, multicollinearity (high correlation between independent variables) should not exist.

### **6. No Outliers or Influential Points**

The data should not contain extreme values (outliers) that can strongly influence the regression line and lead to misleading results.

### **7. Measurement Without Error**

The independent variable (X) is assumed to be measured **without error**, and all errors are assumed to be present only in the dependent variable (Y).

Question 5: What is the method of least squares in linear regression?

Answer:

The **Method of Least Squares** is a statistical technique used in **linear regression** to find the **best-fitting straight line** for a given set of data points. This method determines the values of the regression coefficients such that the difference between the **observed values** and the **predicted values** is minimized.

### **Concept Explanation**

In a simple linear regression model, the equation is:
[
Y = a + bX
]

For each data point, there is a difference between:

* **Actual value (Y)**
* **Predicted value ((\hat{Y}))**

This difference is called the **error or residual**.

### **Core Idea of Least Squares**

The method of least squares minimizes the **sum of the squares of the residuals**, that is:

‚àë(Y‚àíY^)2

Squaring the errors:

* Makes all values positive
* Gives more importance to larger errors
* Helps in finding a unique best-fit line

### **Steps in the Least Squares Method**

1. Assume a straight-line equation (Y = a + bX)
2. Calculate predicted values of Y
3. Find residuals (Y ‚àí ≈∂)
4. Square the residuals
5. Minimize the total of squared residuals to find optimal values of **a** and **b**


### **Purpose of the Method**

1. To obtain the most accurate regression line
2. To minimize prediction errors
3. To ensure the line fits the data as closely as possible


### **Example**

If student marks are predicted using study hours, the least squares method finds the line that best represents the relationship between hours studied and marks obtained.

Below are **three commonly used evaluation metrics** for regression models:

### **1. Mean Absolute Error (MAE)**

MSE=n1‚Äã‚àë(Y‚àíY^)2

* Penalizes larger errors more than smaller ones

* Expressed in the same unit as the dependent variable

Question 8: What is the purpose of the R-squared metric in regression analysis?

Answer:

**R-squared (R¬≤)**, also known as the **coefficient of determination**, is a statistical measure used to evaluate the **goodness of fit** of a regression model. It shows how well the independent variable(s) explain the variation in the dependent variable.


### **Meaning of R-squared**

R-squared represents the **proportion of variance in the dependent variable (Y)** that is explained by the independent variable(s) (X).

Its value lies between **0 and 1**:

* **R¬≤ = 0** ‚Üí Model explains none of the variation
* **R¬≤ = 1** ‚Üí Model explains all the variation


### **Mathematical Expression**

[
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
]

Where:

* (SS_{res}) = Sum of squares of residuals
* (SS_{tot}) = Total sum of squares


### **Purpose of R-squared**

1. **Measures model accuracy**
   Indicates how well the regression line fits the data.

2. **Explains variance**
   Shows the percentage of variation in Y explained by X.

3. **Model comparison**
   Helps compare different regression models; higher R¬≤ indicates better fit.

4. **Evaluates prediction quality**
   Higher R¬≤ generally means better predictions.


### **Example**

If **R¬≤ = 0.85**, it means **85% of the variation** in the dependent variable is explained by the regression model, while 15% is due to other factors or noise.


### **Limitations**

* A high R¬≤ does not guarantee a good model
* Does not indicate causation
* Can increase even with irrelevant variables



Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)

Answer:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# X = independent variable (reshaped as 2D array)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)

# Y = dependent variable
Y = np.array([2, 4, 6, 8, 10])

# Create the model
model = LinearRegression()

# Fit the model
model.fit(X, Y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print results
print("Slope (Coefficient):", slope)
print("Intercept:", intercept)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

Answer:
In a **Simple Linear Regression (SLR)** model, the relationship between the independent variable (X) and the dependent variable (Y) is expressed as:

[
Y = a + bX
]

The model contains **two main coefficients**: the **intercept (a)** and the **slope (b)**. Interpreting these coefficients helps us understand how X affects Y.

### **1. Intercept (a)**

* The intercept represents the **value of the dependent variable (Y) when the independent variable (X) is zero**.
* It shows where the regression line **cuts the Y-axis**.
* In real-world terms, it indicates the **baseline value** of Y without the influence of X.

**Example:**
If the intercept is **20**, it means when X = 0, Y is expected to be 20.

### **2. Slope (b)**

* The slope indicates the **change in the dependent variable (Y) for a one-unit increase in the independent variable (X)**.
* It shows the **direction and strength** of the relationship.

  * **Positive slope:** Y increases as X increases
  * **Negative slope:** Y decreases as X increases
* A larger absolute value of the slope means a **stronger effect** of X on Y.

**Example:**
If the slope is **5**, it means Y increases by 5 units for every 1-unit increase in X.

### **Practical Interpretation**

If a regression equation is:
[
Y = 10 + 3X
]

* Intercept (10): Expected Y when X = 0
* Slope (3): Y increases by 3 units for each additional unit of X
